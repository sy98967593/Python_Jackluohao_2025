# 基础运算
Numpy是一个科学计算模块，用于处理大型多维数组和矩阵，并执行各种矩阵运算。
Numpy合理运用了C语言的能力，将Python的易用性和C的高效完美结合，使得Numpy成为Python科学计算的核心。
而Pandas则提供了DataFrame对象，用于处理数据，如数据导入、数据处理、数据可视化等等。和Numpy的最大区别是，Pandas的DataFrame对象是二维的，而Numpy的数组是多维的。

### 1. 加减乘除
Numpy可以并行运算。
用列表你需要使用到一个循环


```python
# 列表运算
l = [1, 2, 3, 4, 5]
for i in range(len(l)):
    l[i]+=3
print(l)
```

```python
# Numpy运算
import numpy as np
a = np.array([1,2,3])
print("a + 3:", a + 3)
print("a - 3:", a - 3)
print("a * 3:", a * 3)
print("a / 3:", a / 3)
```

```python
# 矩阵乘法
a = np.array([
[1, 2],
[3, 4]
])
b = np.array([
[5, 6],
[7, 8]
])

print(a.dot(b))
print(np.dot(a, b))

    """矩阵还有很多其他的计算，
    比如 np.outer() 矩阵外积，np.inner() 矩阵内积 （和 np.dot() 的用法稍稍有些不同，
    你可以理解成 np.dot(a, b）= np.inner(a, b.T), 把 b 做一次转置）。
    """
```

```python
# 统计
month_salary = [5000, 4000, 3000, 2000, 1000]
print("月工资统计：")
print("均值：", np.mean(month_salary))
print("中位数：", np.median(month_salary))
print("标准差：", np.std(month_salary))
print("最大值：", np.max(month_salary))
print("最小值：", np.min(month_salary))
```

```python
# argmax()和argmin(),用来获取最大值和最小值的索引
a = np.array([150, 166, 183, 170])
name = ["小米", "OPPO", "Huawei", "诺基亚"]
high_idx = np.argmax(a)
low_idx = np.argmin(a)
print("{} 最高".format(name[high_idx]))
print("{} 最矮".format(name[low_idx]))
```
