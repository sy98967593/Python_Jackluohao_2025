# 疫情数据分析

```python
with open('archive/day_wise.csv', 'r', encoding='utf-8') as f:
    data = f.readlines() # 读取所有行

covid = { # 定义一个字典，用于存储数据
    "date": [], # 日期
    "data": [], # 数据
    "header": [h for h in data[0].split(",") [1:]], # 头部
}

for row in data[1:]: # 遍历每一行数据
    split_row = row.strip().split(",") # 按逗号分隔,strip()去除换行符
    covid["date"].append(split_row[0]) # 日期在第一列
    covid["data"].append([float(n) for n in split_row[1:]]) # 数据在第二列开始
```

# 数据清洗

```python
# 原始数据
raw_data = [
["Name", "StudentID", "Age", "AttendClass", "Score"],
["小明", 20131, 10, 1, 67],
["小花", 20132, 11, 1, 88],
["小菜", 20133, None, 1, "98"],
["小七", 20134, 8, 1, 110],
["花菜", 20134, 98, 0, None],
["刘欣", 20136, 12, 0, 12]
]

raw_data
```

 Numpy 面对大数据时，它得天独厚的计算速度优势了。同样一份 1GB 的数据，Python List 处理起来可能要 1 小时， Numpy 优化后，速度可以成倍增长。

```python
import numpy as np
data = np.array(raw_data)
data # 结果里有 dtype=object ，说明数据类型是object，即对象，无法参与Numpy的运算，只有dtype是int，float这种才行。
```

```python
# 数据预处理
data_process = []
for i in range(len(raw_data)):
    if i == 0:
        continue    # 不要首行字符串
    # 去掉首列名字
    data_process.append(raw_data[i][1:])
data = np.array(data_process, dtype=np.float64) # 这里只有float能转换None，int不能。
print("data.dtype", data.dtype)
print(data) 
```

```python
# 数据清洗
sid = data[:,0]# 括号里意思是 获取data的0列，即所有行的第0列，即所有学生的学号。
unique,counts = np.unique(sid,return_counts=True)
print(counts)
print(unique[counts>1]) # 打印出重复的学号
```

```python
# 根据清洗结果修改数值
data[4, 0] = 20135 
print(data)
```

```python
# 清洗年龄
is_nan = np.isnan(data[:,1])
print("is_nan:", is_nan)
nan_idx = np.argwhere(is_nan)

# 计算有数据的平均年龄，用 ~ 符号可以 True/False 对调
mean_age = data[~np.isnan(data[:,1]), 1].mean() # ~ 符号可以 True/False 对调，这里用于筛选数据。
print("有数据的平均年龄：", mean_age)
```
